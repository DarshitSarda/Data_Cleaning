```python
# Step 1: Convert decimal numbers to binary
x = bin(1)[2:]  # Convert 1 to binary and remove the '0b' prefix
y = bin(2)[2:]  # Convert 2 to binary and remove the '0b' prefix

# Step 2: Perform logical OR operation
or_result = bin(int(x, 2) | int(y, 2))[2:]  # Perform OR operation on binary numbers and convert back to binary

# Step 3: Convert binary result back to plain text
or_result_text = chr(int(or_result, 2))  # Convert binary to decimal and then to plain text

print(or_result_text)  # Output the result
```

```python
import pandas as pd
df = pd.read_excel("Customer Call List.xlsx")
```

```python
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns="Not_Useful_Column")
```

```python
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--","")
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--","")
```

```python
df
```

```python
df[['Street_Address','State','Zip_Code']] = df["Address"].str.split(",",n=2,expand=True)
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")

df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
df
```

```python
df = df.replace('N/a',"")
```

```python
df
```

```python
df = df.fillna("")
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x,inplace=True)
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x,inplace=True)
df
```

```python
df = df.reset_index(drop=True)
```
